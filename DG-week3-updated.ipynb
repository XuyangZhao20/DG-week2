```python
date1=[]
for i in range(len(master['Date of Travel'])):
    line = master['Date of Travel'][i]
    d = line.strip().split('/')
    if len(d[2]) == 4:
        year = d[2]
    if len(d[0]) < 2:
        month = '0'+ d[0]
    else:
        month = d[0]
    if len(d[1]) < 2:
        day = '0'+ d[1]
    else:
        day = d[1]
        

    date = year + '/'+ month +'/'+ day 
    date1.append(date)

master['Date of Travel']=date1
master
```

```python
# Time period = 2016/1/31 - 2018/12/31
t=[]
for i in master['Date of Travel']:
    if i >= '2016/01/31':
        if i <='2018/12/31':
            t.append(0)
        else:
            t.append(1)
    else:
        t.append(1)
```

```python
master['index']=t
index1=master[master['index']==1].index
master=master.drop(index1,axis=0)
```

```python
master=master.drop(['index'],axis=1)
master
```

```python
master['Profit'] = master['Price Charged']-master['Cost of Trip']
master
```

```python
master['Price per km']=master['Price Charged']/master['KM Travelled']
master['Cost per km'] = master['Cost of Trip']/master['KM Travelled']
master
```

```python
for i in range(len(city['Population'])):
    city['Population'][i]=int(city['Population'][i].replace(',', ''))
```

```python
# Active Users
active_ratio_2016 = []
active_ratio_2017 = []
active_ratio_2018 = []
active_num_2016 = []
active_num_2017 = []
active_num_2018 = []
set1=master[master['Date of Travel']<='2017/01/01']
set2=master[(master['Date of Travel']>='2017/01/01' )& (master['Date of Travel']<='2018/01/01')]
set3=master[master['Date of Travel']>='2018/01/01']
for i in city['City']:
    temp1 = set1['Customer ID'][master['City']==i]
    n1 = len(set(temp1))
    ratio1 = n1/int(city['Users'][city['City']==i])
    active_ratio_2016.append(ratio1)
    active_num_2016.append(n1)
    
    temp2 = set2['Customer ID'][master['City']==i]
    n2 = len(set(temp2))
    ratio2 = n2/int(city['Users'][city['City']==i])
    active_ratio_2017.append(ratio2)
    active_num_2017.append(n2)
    
    temp3 = set3['Customer ID'][master['City']==i]
    n3 = len(set(temp3))
    ratio3 = n3/int(city['Users'][city['City']==i])
    active_ratio_2018.append(ratio3)
    active_num_2018.append(n3)
    
city['active_ratio_2016'] = active_ratio_2016
city['active_ratio_2017'] = active_ratio_2017
city['active_ratio_2018'] = active_ratio_2018
city['active_users_2016'] = active_num_2016
city['active_users_2017'] = active_num_2017
city['active_users_2018'] = active_num_2018
```

```python
# Hypothesis 2: Is there a gender preference for choosing cab company
cab = ['Yellow', 'Pink']
f = []
m = []
set_y=master[master['Company']=="Yellow Cab"]
set_p=master[master['Company']=="Pink Cab"]
set_01 = set_y['Customer ID'][set_y['Gender']=='Male']
set_02 = set_y['Customer ID'][set_y['Gender']=='Female']
set_03 = set_p['Customer ID'][set_p['Gender']=='Male']
set_04 = set_p['Customer ID'][set_p['Gender']=='Female']
m.append(len(set(set_01)))
f.append(len(set(set_02)))
m.append(len(set(set_03)))
f.append(len(set(set_04)))

plt.figure()
plt.title("Number of Users in Different Gender")
plt.bar(cab, height=m, label = 'Male')
plt.bar(cab, height=f, label = 'Female')
plt.legend()
plt.show()
```

```python
from scipy import stats
```

```python
# Chi Square Test - Gender preference
obs=np.array([m[0],m[1],f[0],f[1]])
stats.chi2_contingency(obs)[0:3]
```

```python
# So p-value = 1 > 0.95, so there is a gender preference of cab company choice among users.
```
